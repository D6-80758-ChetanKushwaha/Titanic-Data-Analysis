```python
from sklearn.linear_model import LogisticRegression
from sklearn.utils import shuffle

df = shuffle(df, random_state=42)
df.head()
```

```python
df.head()
```

```python
df_x = df.drop('Survived',axis=1)
df_y = df['Survived']
```

```python
log_reg = LogisticRegression()
log_reg.fit(df_x,df_y)
print(log_reg.intercept_)
print(log_reg.coef_)
```

```python
test = pd.read_csv('titanic/test.csv',index_col=0)
test = test.drop(columns=['Name','Ticket','Cabin'])
test['Sex'].replace({'male':1,
                  'female':0},inplace=True)
test['Embarked'] = test['Embarked'].astype('category').cat.codes
test = test.replace({-1:3})
test.interpolate(inplace=True)
```

```python
test.head()
```

```python
log_reg.predict(test)
```

```python
lebal = pd.read_csv('titanic/gender_submission.csv',index_col=0)
lebal.head()
```

```python
lebal['pridected'] = log_reg.predict(test)
```

```python
log_reg.score(test,lebal['Survived'])
```
